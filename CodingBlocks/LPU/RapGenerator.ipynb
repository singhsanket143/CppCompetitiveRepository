```python
import numpy as np
from gtts import gTTS
```

```python
# Read the file
def load_text(filename):
    with open(filename, encoding="utf-8") as f:
        return f.read().lower()
```

```python
data = load_text("data.txt")
```

```python
def transitionTable(data, k=4):
    T = {}
    # Work
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        elif T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y]+=1
            
    return T
```

```python
test = transitionTable(data, 4)
```

```python
def probability(T):
    for kx in T.keys(): # This will go to every substring
        norm_factor = sum(T[kx].values()) # Denominator
        for val in T[kx].keys():
            T[kx][val]/=norm_factor
    return T
```

```python
test_probab = probability(test)
```

```python
test_probab
```

```python
def predict(string, T):
    if T.get(string) is None:
        return " "
    next_char = list(T[string].keys())
    probab = list(T[string].values())
    return np.random.choice(next_char, p=probab)
```

```python
sentence = " कौन"
for i in range(1000):
    next_char = predict(sentence[-4:], test_probab)
    sentence+=next_char
print(sentence)
```

```python
file = gTTS(sentence, lang="hi")
file.save("test.mp3")
```
